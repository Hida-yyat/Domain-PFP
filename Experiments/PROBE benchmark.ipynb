## Embeddings

Our generated embeddings can be found in


* ***../data/embeddings/PROBE_embeddings.csv*** (trained on entire SwissProt except the human proteins)
* ***../data/embeddings/PROBE_embeddings_75.csv*** (trained on entire SwissProt with no more than 75% sequence identity with the human proteins)
* ***../data/embeddings/PROBE_embeddings_50.csv*** (trained on entire SwissProt with no more than 50% sequence identity with the human proteins)
* ***../data/embeddings/PROBE_embeddings_25.csv*** (trained on entire SwissProt with no more than 25% sequence identity with the human proteins)
